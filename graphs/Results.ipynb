```python
# graph NIS from laser and radar

import matplotlib.pyplot as plt
import numpy as np

# Get data for both datasets
laser_path_data1 = "../output/laserOutput_Dataset1.txt"
radar_path_data1 = "../output/radarOutput_DataSet1.txt"
laser_path_data2 = "../output/laserOutput_Dataset2.txt"
radar_path_data2 = "../output/radarOutput_DataSet2.txt"

# arrays to hold each dataset
lidar_data1 = []
radar_data1 = []
lidar_data2 = []
radar_data2 = []
```

```python
with open(laser_path_data1) as f:
    line = f.readline()
    
    while line:
        #print("{}".format(line.strip()))
        
        lidar_data1.append(float(line))
        line = f.readline()
    
f.close()

lidar_len1 = len(lidar_data1)
    
```

```python
with open(laser_path_data2) as f:
    line = f.readline()
    
    while line:
        #print("{}".format(line.strip()))
        
        lidar_data2.append(float(line))
        line = f.readline()
    
f.close()

lidar_len2 = len(lidar_data2)
```

```python
with open(radar_path_data1) as f:
    line = f.readline()
    
    while line:
        #print("{}".format(line.strip()))
        
        radar_data1.append(float(line))
        line = f.readline()
    
f.close()

radar_len1 = len(radar_data1)
```

```python
with open(radar_path_data2) as f:
    line = f.readline()
    
    while line:
        #print("{}".format(line.strip()))
        
        radar_data2.append(float(line))
        line = f.readline()
    
f.close()

radar_len2 = len(radar_data2)
```

```python
fig = plt.figure(figsize= (20,5))

plt.subplot(1,2,1)
plt.plot(lidar_data1, color='g', linestyle='-')
# plot 95% line for 2DOF = values higher than 5.991
plt.axhline(y=5.991, color='r', linestyle='-', label='2DOF NIS = 5.991')
plt.axis([0, lidar_len1, 0, 15])
plt.xlabel('k')
plt.ylabel('ε')
plt.title('Laser Dataset1 NIS Results')
plt.legend(loc="upper right")

plt.subplot(1,2,2)
plt.plot(lidar_data2, color='g', linestyle='-')
# plot 95% line for 2DOF = values higher than 5.991
plt.axhline(y=5.991, color='r', linestyle='-', label='2DOF NIS = 5.991')
plt.axis([0, lidar_len2, 0, 15])
plt.xlabel('k')
plt.ylabel('ε')
plt.title('Laser Dataset2 NIS Results')
plt.legend(loc="upper right")

plt.savefig("../support/LaserNIS_Graph.png")
plt.show()

```

```python
fig = plt.figure(figsize= (20,5))

plt.subplot(1,2,1)
plt.plot(radar_data1, color='c', linestyle='-')
# plot 95% line for 3DOF = values higher than 7.815
plt.axhline(y=7.815, color='r', linestyle='-', label='3DOF NIS = 7.815')
plt.axis([0, radar_len1, 0, 15])
plt.xlabel('k')
plt.ylabel('ε')
plt.title('Radar Dataset1 NIS Results')
plt.legend(loc="upper right")

plt.subplot(1,2,2)
plt.plot(radar_data2, color='c', linestyle='-')
# plot 95% line for 3DOF = values higher than 7.815
plt.axhline(y=7.815, color='r', linestyle='-', label='3DOF NIS = 7.815')
plt.axis([0, radar_len2, 0, 15])
plt.xlabel('k')
plt.ylabel('ε')
plt.title('Radar Dataset2 NIS Results')
plt.legend(loc="upper right")

plt.savefig("../support/RadarNIS_Graph.png")
plt.show()

```
