```python
# graph NIS from laser and radar

import matplotlib.pyplot as plt
import numpy as np

laser_path = "../output/laserOutput_Dataset1.txt"
radar_path = "../output/radarOutput_DataSet1.txt"



lidar_data = []
radar_data = []
```

```python
with open(laser_path) as f:
    line = f.readline()
    
    while line:
        #print("{}".format(line.strip()))
        
        lidar_data.append(float(line))
        line = f.readline()
    
f.close()

lidar_len = len(lidar_data)
    
```

```python
plt.plot(lidar_data, 'ro')
# plot 95% line for 2DOF = values higher than 5.991
plt.axhline(y=5.991, color='b', linestyle='-')
plt.axis([0, lidar_len, 0, 20])

plt.title('Laser')
plt.show()
```

```python
with open(radar_path) as f:
    line = f.readline()
    
    while line:
        #print("{}".format(line.strip()))
        
        radar_data.append(float(line))
        line = f.readline()
    
f.close()

radar_len = len(radar_data)
        
```

```python
#plt.plot(radar_data, 'bo')
plt.plot(radar_data, color='g', linestyle='-')
# plot 95% line for 3DOF = values higher than 7.815
plt.axhline(y=7.815, color='r', linestyle='-')
plt.axis([0, radar_len, 0, 20])
plt.title('Radar')
plt.show()
```
